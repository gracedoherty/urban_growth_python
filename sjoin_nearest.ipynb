```python
#df1 = gpd.read_file("https://github.com/gracedoherty/urban_growth/blob/main/df1.zip?raw=true")
#df2 = gpd.read_file("https://github.com/gracedoherty/urban_growth/blob/main/df2.zip?raw=true")
df1 = gpd.read_file("povertyequity/urban_growth/WSFE_ADM3_ply.shp")
df2 = gpd.read_file("povertyequity/urban_growth/GRID3_ADM3_ply.shp")
```

```python
df1.sample(10)
```

```python
df1.reset_index(inplace=True)
df1.rename(columns={'index': 'WSFE_ID'}, inplace=True)
df1 = df1.drop(columns=['Id', 'gridcode', 'Shape_Area'])
df1.sample(10)
```

```python
df2.sample(10)
```

```python
df2.reset_index(inplace=True)
df2.rename(columns={'index': 'GRID3_splitID', 'GRID3_OBJI':'GRID3_origID'}, inplace=True)
df2 = df2.drop(columns=['Id', 'gridcode', 'Shape_Area', 'gridstring'])
df2.sample(10)
```

```python
# shard the smaller dataframe into a dict
shards = {k:d for k, d in df2.groupby("ADM3_OBJID")}
```

```python
# now just group by ADM, sjoin_nearest appropriate shard
df_n = df1.groupby("ADM3_OBJID").apply(lambda d: gpd.sjoin_nearest(d, shards[d["ADM3_OBJID"].values[0]], max_distance=500))
```

```python
df_n.sample(5, random_state=42)
```

```python
df_n.to_file("CMN_WSFE_GRID3.gpkg", layer="WSFE_GRID3_NearSharding", driver="GPKG")
```
